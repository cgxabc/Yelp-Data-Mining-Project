```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
dish_counts={}
top_count_index=[]
with open('dish_to_counts') as f:
    for line in f.readlines():
        line=line.rstrip()
        line = line.split("\t")
        #print line
        name = line[0]
        top_count_index.append(name)
        dish_counts[name]=int(line[1])
dish_counts
```

```python
dish_score={}
top_score_index=[]
with open('dish_to_score') as f:
    for line in f.readlines():
        line=line.rstrip()
        line = line.split("\t")
        #print line
        name = line[0]
        top_score_index.append(name)
        dish_score[name]=float(line[1])
dish_score
```

```python
top_count_index[0:14]
```

```python
top_50_dish_count=pd.Series(dish_counts,top_count_index[0:50])
```

```python
top_20_dish_count=pd.Series(dish_counts,top_count_index[0:20])
```

```python
top_50_dish_score=pd.Series(dish_score,top_score_index[0:50])
```

```python
top_20_dish_score=pd.Series(dish_score,top_score_index[0:20])
```

```python
plt.figure(figsize=(12,8))
plt.barh(np.arange(len(top_20_dish_count)),top_20_dish_count.values,align='center',alpha=0.5,color='purple')
plt.yticks(np.arange(len(top_20_dish_count)),top_20_dish_count.index,fontsize=15)
plt.xlabel('Counts in reviews')
plt.title("Top 20 most mentioned Chinese dishes",fontsize=20)
```

```python
plt.figure(figsize=(12,8))
plt.bar(np.arange(len(top_50_dish_count)),top_50_dish_count.values,color='orange')
plt.xticks(np.arange(len(top_50_dish_count)),top_50_dish_count.index,rotation=90,fontsize=12)
plt.ylabel('Counts in reviews')
plt.title("Top 50 most mentioned Chinese dishes",fontsize=20)
```

```python
plt.figure(figsize=(12,8))
plt.barh(np.arange(len(top_20_dish_score)),top_20_dish_score.values,align='center',alpha=0.5,color='blue')
plt.yticks(np.arange(len(top_20_dish_score)),top_20_dish_score.index,fontsize=15)
plt.xlabel('Scores in reviews')
plt.title("Top 20 most popular Chinese dishes",fontsize=20)
```

```python
plt.figure(figsize=(12,8))
plt.bar(np.arange(len(top_50_dish_score)),top_50_dish_score.values,color='green')
plt.xticks(np.arange(len(top_50_dish_score)),top_50_dish_score.index,rotation=90,fontsize=12)
plt.ylabel('Scores in reviews')
plt.title("Top 50 most popular Chinese dishes",fontsize=20)
```

```python
dim_sum_rest_score={}
dim_sum_index=[]
with open('dim_sum_rest2score') as f:
    for line in f.readlines():
        line = line.rstrip()
        line=line.split("\t")
        dim_sum_index.append(line[0])
        dim_sum_rest_score[line[0]]=float(line[1].rstrip())
dim_sum_rest_score
```

```python
top_50_dim_sum=pd.Series(dim_sum_rest_score,dim_sum_index[0:50])
```

```python
plt.figure(figsize=(12,8))
plt.bar(np.arange(len(top_50_dim_sum)),top_50_dim_sum.values,color='darkorange')
plt.xticks(np.arange(len(top_50_dim_sum)),top_50_dim_sum.index,rotation=90,fontsize=10)
plt.ylabel('Scores')
plt.title("Top 50 most popular restaurants in dim sum",fontsize=20)
```
